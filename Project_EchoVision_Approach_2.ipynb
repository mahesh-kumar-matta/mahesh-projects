```python
from google import genai
from google.genai import types
from transformers import pipeline
from IPython.display import Audio as IPythonAudio

f = open("keys/.gemini.txt")
key = f.read()

client = genai.Client(api_key=key)

with open('images/image_1.JPEG', 'rb') as f:
  image_bytes = f.read()

context = [
    types.Part.from_bytes(
        data=image_bytes,
        mime_type='image/jpeg',
    ),
]

# Below is for Step-1 requirement
#custom_config = {
#    "system_instruction" : "You are a helpful AI Assistant. Given an image perform object detection and provide a text output which contains the information about the labels detected and their counts.",
#}

# Step-2 : Enhancement
custom_config = {
    "system_instruction" : "You are a helpful AI Assistant. Given an image Prepare the generated caption for speech synthesis. Remove unnecessary symbols. Provide the output text in detail. Provide text in polite tone. Start the text out with.. Here is what I see in the image..",
}

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    config=custom_config,
    contents=context,
)

result_text = response.text

print(result_text)

# Step-3: Generate speech from text

# create object for text-to-speech
narrator = pipeline(
    task="text-to-speech",
    model="facebook/mms-tts-eng"
)

# convert from text to speech
narrated_text = narrator(result_text)

# Play the audio
IPythonAudio(
    narrated_text["audio"],
    rate=narrated_text["sampling_rate"]
)
```
